```python
# import hierarchical clustering libraries
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
```

```python
data=pd.read_csv("C:\\Users\\DELL\\Downloads\\EastWestAirline_new.csv")
data.head()
```

```python
data.isnull().sum()
# No null values
```

```python
df=data.drop(["ID#","cc2_miles","cc3_miles"],axis=1)
df.head()
```

```python
df.isnull().sum()
# No null values
```

```python
sns.pairplot(df)
```

```python
sns.distplot(df)
```

```python
df.skew()
```

```python
# Normalization function 
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
# Normalized data frame (considering the numerical part of data)
df_norm = norm_func(df)
df_norm.head()
```

```python
kmeans_clust=KMeans(n_clusters=2)
kmeans_clust.fit(df_norm)
kmeans_clust.labels_
```

```python
k = list(range(2,8))
TWSS = []
for i in k:
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(df_norm)
    WSS = []
    for j in range(i):
        WSS.append(sum(cdist(df_norm.iloc[kmeans.labels_==j,:],kmeans.cluster_centers_[j].reshape(1,df_norm.shape[1]),"euclidean")))
    TWSS.append(sum(WSS))
TWSS
```

```python
plt.plot(k,TWSS,"ro-")
```

```python
kmeans_clust=KMeans(n_clusters=5)
kmeans_clust.fit(df_norm)
kmeans_clust.labels_
```

```python
df_norm["clusters"]=pd.Series(kmeans_clust.labels_)
```

```python
result=df.groupby(df_norm.clusters).mean()
result
```
